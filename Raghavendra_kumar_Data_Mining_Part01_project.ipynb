```python
#Clustering: Treat missing values in CPC, CTR and CPM using the formula given.
```

```python
# Calculate CPM, CPC and CTR for rows where they are NaN
ads_data.loc[ads_data['CPM'].isnull(), 'CPM'] = (ads_data['Spend'] / ads_data['Impressions']) * 1000
ads_data.loc[ads_data['CPC'].isnull(), 'CPC'] = ads_data['Spend'] / ads_data['Clicks']
ads_data.loc[ads_data['CTR'].isnull(), 'CTR'] = (ads_data['Clicks'] / ads_data['Impressions']) * 100
```

```python
#Clustering: Check if there are any outliers. Do you think treating outliers is necessary for K-Means clustering? Based on your judgement decide whether to treat outliers and if yes, which method to employ. (As an analyst your judgement may be different from another analyst).
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Boxplot for CPM
sns.boxplot(x=ads_data['CPM'], ax=ax[0])
ax[0].set_title('Boxplot for CPM')

# Boxplot for CPC
sns.boxplot(x=ads_data['CPC'], ax=ax[1])
ax[1].set_title('Boxplot for CPC')

# Boxplot for CTR
sns.boxplot(x=ads_data['CTR'], ax=ax[2])
ax[2].set_title('Boxplot for CTR')

plt.tight_layout()
plt.show()
```

```python
# Columns to treat outliers for
columns_to_treat = ['CPM', 'CPC', 'CTR']

# Capping outliers using the IQR method
for column in columns_to_treat:
    Q1 = ads_data[column].quantile(0.25)
    Q3 = ads_data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Defining the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Capping the values outside the bounds
    ads_data[column] = ads_data[column].clip(lower=lower_bound, upper=upper_bound)

# Boxplots after treating outliers
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Boxplot for CPM after treating outliers
sns.boxplot(x=ads_data['CPM'], ax=ax[0])
ax[0].set_title('Boxplot for CPM after treating outliers')

# Boxplot for CPC after treating outliers
sns.boxplot(x=ads_data['CPC'], ax=ax[1])
ax[1].set_title('Boxplot for CPC after treating outliers')

# Boxplot for CTR after treating outliers
sns.boxplot(x=ads_data['CTR'], ax=ax[2])
ax[2].set_title('Boxplot for CTR after treating outliers')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Extracting numeric columns for scaling
numeric_columns = ads_data.select_dtypes(include=['float64', 'int64']).columns

# Performing z-score scaling
scaler = StandardScaler()
ads_data_scaled = ads_data.copy()
ads_data_scaled[numeric_columns] = scaler.fit_transform(ads_data[numeric_columns])
ads_data_scaled.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Using the WARD method and Euclidean distance for hierarchical clustering
linked = linkage(ads_data_scaled[numeric_columns].sample(1000, random_state=42), method='ward', metric='euclidean')

# Plotting the dendrogram
plt.figure(figsize=(15, 10))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram (using WARD)')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
#Clustering: Make Elbow plot (up to �=10n=10) and identify the optimum number of clusters for k-means algorithm
```

```python
from sklearn.cluster import KMeans
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
# List to store the sum of squared distances
ssd = []

# Range of clusters
clusters = range(1, 11)

# Calculating SSD for each number of clusters
for cluster in clusters:
    kmeans = KMeans(n_clusters=cluster, random_state=42)
    kmeans.fit(ads_data_scaled[numeric_columns])
    ssd.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(clusters, ssd, '-o', markerfacecolor='red', markersize=8)
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Distances (SSD)')
plt.grid(True)
plt.show()
```

```python
# Clustering: Print silhouette scores for up to 10 clusters and identify the optimum number of clusters.
```

```python
from sklearn.metrics import silhouette_score
```

```python
# List to store silhouette scores
sil_scores = []

# Calculating silhouette score for each number of clusters
for cluster in clusters:
    kmeans = KMeans(n_clusters=cluster, random_state=42)
    cluster_labels = kmeans.fit_predict(ads_data_scaled[numeric_columns])
    
    # We don't calculate silhouette score for 1 cluster since it's not meaningful
    if cluster > 1:
        score = silhouette_score(ads_data_scaled[numeric_columns], cluster_labels)
        sil_scores.append(score)
    else:
        sil_scores.append(0)

# Plotting silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(clusters, sil_scores, '-o', markerfacecolor='green', markersize=8)
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
sil_scores
```

```python
# Clustering: Profile the ads based on optimum number of clusters using silhouette score and your domain understanding [Hint: Group the data by clusters and take sum or mean to identify trends in Clicks, spend, revenue, CPM, CTR, & CPC based on Device Type. Make bar plots].
```

```python
# K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
ads_data['Cluster'] = kmeans.fit_predict(ads_data_scaled[numeric_columns])

# Confirm the addition of cluster labels
ads_data[['Cluster']].head()
```

```python
# Grouping by clusters to profile the data
cluster_summary = ads_data.groupby('Cluster').agg({
    'Clicks': ['mean', 'sum'],
    'Spend': ['mean', 'sum'],
    'Revenue': ['mean', 'sum'],
    'CPM': 'mean',
    'CTR': 'mean',
    'CPC': 'mean'
}).reset_index()

cluster_summary.columns = ['Cluster', 'Clicks_Mean', 'Clicks_Sum', 'Spend_Mean', 'Spend_Sum', 'Revenue_Mean', 
                           'Revenue_Sum', 'CPM_Mean', 'CTR_Mean', 'CPC_Mean']

# Device Type breakdown
device_type_counts = ads_data.groupby(['Cluster', 'Device Type']).size().unstack().reset_index()

# Merging the device type counts with the summary statistics
cluster_profile = cluster_summary.merge(device_type_counts, on='Cluster')

# Displaying the cluster profile
cluster_profile
```

```python
# Redefining the metrics for visualization
metrics_to_plot = ['Clicks_Mean', 'Spend_Mean', 'Revenue_Mean', 'CPM_Mean', 'CTR_Mean', 'CPC_Mean']

# Visualizing the profiles using bar plots again

fig, ax = plt.subplots(len(metrics_to_plot), 1, figsize=(10, 15))

for i, metric in enumerate(metrics_to_plot):
    sns.barplot(x='Cluster', y=metric, data=cluster_profile, ax=ax[i], palette='viridis')
    ax[i].set_title(metric.replace('_', ' '))
    ax[i].set_ylabel('')
    ax[i].set_xlabel('Cluster')

plt.tight_layout()
plt.show()
```
