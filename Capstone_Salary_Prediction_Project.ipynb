```python
categorical_columns = ['Department', 'Role', 'Industry', 'Organization', 'Designation', 'Graduation_Specialization',
                       'University_Grad', 'PG_Specialization', 'University_PG', 'PHD_Specialization', 'University_PHD',
                       'Last_Appraisal_Rating']

for column in categorical_columns:
    data[column] = data[column].fillna('Unknown')
```

```python
# Outlier treatment
```

```python
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower=lower_bound, upper=upper_bound)

data['Current_CTC'] = cap_outliers(data['Current_CTC'])
data['Expected_CTC'] = cap_outliers(data['Expected_CTC'])

data.describe()
```

```python
# Select features and target variable
features = ['Total_Experience', 'Total_Experience_in_field_applied', 'Current_CTC']
target = 'Expected_CTC'

X = data[features]
y = data[target]
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize Linear Regression model
linear_regression = LinearRegression()

# Train the model
linear_regression.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the testing set
y_pred = linear_regression.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display performance metrics
print("Linear Regression Model Performance:")
print(f"MSE: {mse}")
print(f"R-squared: {r2}")
```
